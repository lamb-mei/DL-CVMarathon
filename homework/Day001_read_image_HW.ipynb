思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度


假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

```python
import cv2
from pathlib import Path
import os
import sys
from  matplotlib import pyplot as plt

print(sys.argv[0])
print(sys.argv)
print(os.getcwd())
print(Path().resolve())

def getPath(path):
    imgPath = None
    cwd = os.path.abspath('')
    path = r"%s/%s" % (cwd , path)
    imgPath = Path(path).resolve()

    imgPath = str(imgPath)
    return imgPath

img_path = './data/lena.png'
img_path = getPath(img_path)
print(img_path)
img_bgrimg_bgr = cv2.imread(img_path, cv2.IMREAD_COLOR)
# B, G, R = cv2.split(img_bgrimg_bgr)
_B = img_bgrimg_bgr[:, :, 0]
_G = img_bgrimg_bgr[:, :, 1]
_R = img_bgrimg_bgr[:, :, 2]

# cv2.imshow('B', _B)
# cv2.imshow('G', _G)  
# cv2.imshow('R', _R)
# cv2.waitKey(0)


b = img_bgrimg_bgr.copy()
# set green and red channels to 0
b[:, :, 1] = 0
b[:, :, 2] = 0

g = img_bgrimg_bgr.copy()
# set blue and red channels to 0
g[:, :, 0] = 0
g[:, :, 2] = 0

r = img_bgrimg_bgr.copy()
# set blue and green channels to 0
r[:, :, 0] = 0
r[:, :, 1] = 0


# cv2.imshow('B-RGB', b)
# cv2.imshow('G-RGB', g)
# cv2.imshow('R-RGB', r)
# cv2.waitKey(0)

cv2.destroyAllWindows()
for i in range (1,4):
    cv2.waitKey(1)


plt.imshow(r)
plt.show()
plt.imshow(g)
plt.show()
plt.imshow(b)
plt.show()
```
